```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('../data_result/final_results_1125.csv')
plt.rcParams.update({'font.size': 12})
```

```python
# First Plot: Overall comparison of the distribution of the three scores between our method and the naive method

# Create a figure with three subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 7))

# List of metrics to analyze
metrics = ['Coherence_score', 'Engagement_score', 'Empathy_score']

# Define a consistent color palette for 'yes', 'no', 'tie'
color_palette = {'yes': '#66c2a5', 'no': '#fc8d62', 'tie': '#8da0cb'}

# Iterate over each metric to create individual bar plots
for i, metric in enumerate(metrics):
    # Get counts of 'yes', 'no', 'tie' for the current metric
    counts = df[metric].value_counts().reindex(['yes', 'no', 'tie'])
    # Create a bar plot for the current metric
    sns.barplot(
        x=counts.index, y=counts.values, ax=axes[i],
        palette=[color_palette.get(x, '#a6d854') for x in counts.index]
    )
    # Set title and labels for the subplot
    axes[i].set_title(f'Distribution of {metric}', fontsize=16)
    axes[i].set_xlabel('Comparison Result', fontsize=14)
    axes[i].set_ylabel('Count', fontsize=14)
    # Increase tick label font size
    axes[i].tick_params(axis='x', labelsize=12)
    axes[i].tick_params(axis='y', labelsize=12)

# Adjust layout to prevent overlap
plt.tight_layout()
# Save the current figure to a PDF file
plt.savefig("Figures/overall_result.pdf", format='pdf', bbox_inches='tight')
# Display the plot
plt.show()
```

```python
# Second Plot: Distribution of the emotions used in the 100 questions

# Get counts of each emotion used in responses, sorted by count in descending order
emotion_counts = df['Emotions'].value_counts().sort_values(ascending=False)

# Create a bar plot for emotion distribution
plt.figure(figsize=(12, 8))
sns.barplot(x=emotion_counts.index, y=emotion_counts.values, palette='Set3')
# Set title and labels for the plot
plt.title('Distribution of Emotions Used in Responses', fontsize=18)
plt.xlabel('Emotion', fontsize=16)
plt.ylabel('Count', fontsize=16)
# Rotate x-axis labels for better readability
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=14)
# Adjust layout
plt.tight_layout()
plt.savefig("Figures/emotion_usage.pdf", format='pdf', bbox_inches='tight')
# Display the plot
plt.show()
```

```python
# Third Plot: Distribution of comparison results grouped by emotion

# List of emotions to consider
emotions_list = ["batman", 'joker', 'honest', 'dishonest', 'happiness', 'sadness',
                 'anger', 'fear', 'disgust', 'surprise']

# Initialize an empty DataFrame to store counts
emotion_metric_counts = pd.DataFrame()

# Iterate over each metric to compute counts grouped by emotion and comparison result
for metric in metrics:
    # Group by 'Emotions' and the current metric, then count occurrences
    temp = df.groupby(['Emotions', metric]).size().reset_index(name='Count')
    # Add a column to indicate the metric
    temp['Metric'] = metric
    # Rename the metric column to 'Comparison_Result' for clarity
    temp.rename(columns={metric: 'Comparison_Result'}, inplace=True)
    # Append to the main DataFrame
    emotion_metric_counts = pd.concat([emotion_metric_counts, temp], ignore_index=True)

# Ensure 'Emotions' is ordered by total counts (as in the second plot)
emotion_order = emotion_counts.index.tolist()
emotion_metric_counts['Emotions'] = pd.Categorical(
    emotion_metric_counts['Emotions'], categories=emotion_order, ordered=True
)

# Create a categorical plot with facets for each metric
g = sns.catplot(
    data=emotion_metric_counts, kind='bar',
    x='Emotions', y='Count', hue='Comparison_Result',
    col='Metric', col_wrap=3,
    height=5, aspect=1.5,
    palette=color_palette
)

# Set titles and labels for the facets
g.set_titles('{col_name}', size=16)
g.set_xticklabels(rotation=45, fontsize=12)
g.set_ylabels('Count', fontsize=14)
g.set_xlabels('Emotion', fontsize=14)
g.set_yticklabels(fontsize=12)
# Set legend title and increase font size
g._legend.set_title('Comparison Result', prop={'size': 14})
for text in g._legend.texts:
    text.set_fontsize(12)
# Adjust layout
plt.tight_layout()
plt.savefig("Figures/emotion_score.pdf", format='pdf', bbox_inches='tight')
# Display the plot
plt.show()
```
